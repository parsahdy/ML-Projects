### multi layer perceptron

```python
import numpy as np
import pandas as pd
```

```python
class MLP:
    def __init__(self, input_size, hidden_size, output_size, learning_rate=0.01, n_epochs=1000):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.learning_rate = learning_rate
        self.n_epochs = n_epochs


        self.weights_input_hidden = np.random.randn(self.input_size, self.hidden_size) * 0.01
        self.weights_hidden_output = np.random.randn(self.hidden_size, self.output_size) * 0.01
        self.bias_hidden = np.zeros((1, self.hidden_size))
        self.bias_output = np.zeros((1, self.output_size))


    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))


    def sigmoid_derivative(self, x):
        return x * (1 - x)


    def fit(self, X, y):
        for epoch in range(self.n_epochs):
            hidden_input = np.dot(X, self.weights_input_hidden) + self.bias_hidden
            hidden_output = self.sigmoid(hidden_input)

            final_input = np.dot(hidden_output, self.weights_hidden_output) + self.bias_output
            final_output = self.sigmoid(final_input)

            error = y - final_output

            d_output = error * self.sigmoid_derivative(final_output)
            error_hidden_layer = d_output.dot(self.weights_hidden_output.T)
            d_hidden = error_hidden_layer * self.sigmoid_derivative(hidden_output)

            
            self.weights_input_hidden += X.T.dot(d_hidden) * self.learning_rate
            self.bias_hidden += np.sum(d_hidden, axis=0, keepdims=True) * self.learning_rate
            
            self.weights_hidden_output += hidden_output.T.dot(d_output) * self.learning_rate
            self.bias_output += np.sum(d_output, axis=0, keepdims=True) * self.learning_rate

            if epoch % 100 == 0:
                loss = np.mean(np.square(error))
                print(f"We had {loss} loss in {epoch} epochs.")


    def predict(self, X):
        hidden_input = np.dot(X, self.weights_input_hidden) + self.bias_hidden
        hidden_output = self.sigmoid(hidden_input)

        final_input = np.dot(hidden_output, self.weights_hidden_output) + self.bias_output
        final_output = self.sigmoid(final_input)
        
        return np.round(final_output)        
        
```

##### Training Data

```python
X = df.select_dtypes(include=['float64', 'int64'] ).values
y = df.iloc[:, -1].values

X = (X - X.min(axis=0)) / (X.max(axis=0)) - (X.min(axis=0))

train_size = int(0.8 * X.shape[0])
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size].reshape(-1, 1), y[train_size:].reshape(-1, 1)


input_size = X_train.shape[1]
hidden_size = 10
output_size = 1

mlp = MLP(input_size=input_size, hidden_size=hidden_size, output_size=output_size)
mlp.fit(X_train, y_train)

predictions = mlp.predict(X_test)
accuracy = np.mean(predictions == y_test)
print(f"Accuracy : {accuracy}")
```
